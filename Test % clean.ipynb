```python
import pandas as pd

raw = pd.read_csv("data/raw/household_power_consumption.txt", sep=";")  
clean = pd.read_csv("data/processed/cleaned_data.csv")

# 1) Tỷ lệ dòng giữ lại
percent_kept = len(clean) / len(raw) * 100
print(f"Giữ lại: {percent_kept:.2f}% (loại bỏ {100-percent_kept:.2f}%)")

# 2) Tỷ lệ giá trị không còn missing
raw_missing = raw.isna().mean().mean() * 100
clean_missing = clean.isna().mean().mean() * 100
print(f"Missing trung bình: trước {raw_missing:.2f}%  |  sau {clean_missing:.2f}%")
```
